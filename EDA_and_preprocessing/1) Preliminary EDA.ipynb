## 1) What is the distribution of the target variable?

Getter function for histogram and boxplot.

```python
# create subplots with 1 row and 2 columns
fig, axs = plt.subplots(ncols=2, figsize=(15,5))

# plot distplot on the first subplot
sns.histplot(data=df, x='target', kde=True, ax=axs[0])
axs[0].set_title('Distribution of target(y)')
# set the x-axis label
axs[0].set_xlabel("target (y)")

# plot boxplot on the second subplot
sns.boxplot(data=df, x='target', ax=axs[1])
axs[1].set_title('Boxplot of target(y)')
axs[1].set_xlabel("target (y)")

# adjust the spacing between subplots
plt.subplots_adjust(wspace=0.3)

# show the plots
plt.show()
```

Observations
* The distribution of the y is skewed to the left.
* There are quite a lot of anomalies from the boxplot on the left.

## 3) What is the distribution like if we log the y distribution?

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# create subplots with 1 row and 2 columns
fig, axs = plt.subplots(ncols=2, figsize=(15,5))

# plot distplot on the first subplot with logged y-axis
sns.histplot(data=df, x='target', kde=True, ax=axs[0])
axs[0].set_title('Distribution of target(y)')
axs[0].set_xlabel("target (y)")
axs[0].set_ylabel("Count (log scale)")
axs[0].set_yscale('log')  # set logarithmic scale for y-axis

# plot boxplot on the second subplot with logged y-axis
sns.boxplot(data=df, x='target', ax=axs[1])
axs[1].set_title('Boxplot of target(y)')
axs[1].set_xlabel("Value (log scale)")
axs[1].set_xscale('log')  # set logarithmic scale for y-axis

# adjust the spacing between subplots
plt.subplots_adjust(wspace=0.3)

# show the plots
plt.show()
```

Observations
* It is now less skewed to the left.
* The log transformation of the y target will be done later in the feature selection python notebook.
